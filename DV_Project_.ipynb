```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
pd.__version__
```

```python
df = pd.read_excel (r'DV_Project.xlsm', sheet_name='Sheet2')
print (df)
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df[['Ethereum','Tether','XRP','Chainlink','Bitcoin Cash','Binance Coin','Litecoin','Polkadot','Cardano','Bitcoin SV']]
Y = df[['USD Coin','EOS','Monero','Wrapped Bitcoin','TRON','Stellar','Tezos','Crypto.com Coin','NEM']]
```

```python
print(X,Y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
print(reg.coef_)
print(reg.intercept_)
```

```python
reg.fit(x_test,y_test)
print(reg.coef_)
print(reg.intercept_)
```

```python
y_pred = reg.predict(x_train)
print('R2 score:', metrics.r2_score(y_train, y_pred))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_train, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_train, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_train, y_pred)))
```

```python
y_pred = reg.predict(x_test)
print('R2 score:', metrics.r2_score(y_test, y_pred))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
